# Add metadata to STM

The Zarr file used in this exmaple can be downloaded [here](https://figshare.com/ndownloader/files/41957250). The file need to be unzipped.

```python
import xarray as xr
import stmtools
```

```python
stmat = xr.open_zarr('./stm.zarr')
stmat
```

```python
stmat.stm.numPoints
```

```python
stmat.stm.numEpochs
```

```python
# Normal registration
metadata_normal = dict(techniqueId='ID0001', datasetId='ID_datasetID', crs=4326)
stmat = stmat.stm.register_metadata(metadata_normal)

# # unnormal registration
# metadata_unnormal = dict(techniqueId=11)
# stmat = stmat.stm.register_metadata(metadata_unnormal)

stmat
```

```python
stmat = stmat.stm.register_datatype(['amplitude', 'phase'], 'obsData') # Register multiple fields 
# stmat = stmat.stm.register_datatype('amplitude', 'obsData') # Register one field

stmat
```

```python
stmat.to_zarr('stm_with_metadata.zarr')
```
